```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
penguins = sns.load_dataset("penguins")
penguins.head()
```

```python
sns.lineplot(x = "bill_length_mm", y = "body_mass_g", hue = "sex", palette = ['blue','orange'], data = penguins)
plt.title('Mass of penguins by bill length')
plt.xlabel('Bill length / mm')
plt.ylabel('Body mass / g')
```

```python
sns.relplot(x = "bill_length_mm", y = "body_mass_g", col = "island", hue = "sex", palette = ['blue','orange'], data = penguins)
plt.xlabel('Bill length / mm')
plt.ylabel('Body mass / g')
```

```python
sns.jointplot(data = penguins, x = 'bill_length_mm', y = 'bill_depth_mm', kind = 'kde')
```

```python
sns.relplot(
    data = penguins,
    x = 'flipper_length_mm',
    y = 'bill_length_mm',
    hue = "island"
#    hue=iris.species.tolist()    <--- This gets past an error with matplotlib 3.3.1!
).set(
    title = "Bill length by flipper length",
    xlabel = 'Flipper Length / mm',
    ylabel = 'Bill Length / mm')
```

```python
peng = penguins.groupby('species').mean()
peng
```
